```python
%matplotlib inline

import numpy as np   
from numpy.random import normal
from numpy.random import uniform 

from nw1d import KernelRegression
```

```python
"""
data_generator(num = 100, scale = 10): Generate a curve with noise on sphere
Input:
    'num': the number of points
    'scale': std for errors
Output:
    a ndarray with shape=(num,3) 
"""
def data_generator(num = 100, scale = 10):
    output = np.empty(shape = (num, 3), dtype="float32")
    x = uniform(low=-1, high=1, size=1) #initial point
    y,z = uniform(low=-1, high=1, size=2)
    d1,d2,d3 = uniform(low=-1, high=1, size=3) #direction
    for i in range(num):
        x = x + d1 + normal(scale = scale)
        y = y + d2 + normal(scale = scale)
        z = z + d3 + normal(scale = scale)
        point = np.array([[x,y,z]])/np.sqrt(x**2 + y**2 + z**2)
        output[i,:] = point 
    return output

#Generate 30 points on the sphere
num = 30
y = data_generator(num, scale=2)
X = np.array(range(num),dtype="float32")
```

```python
model = KernelRegression(dist='sphere')
model.fit(X, y)

x = np.array([5.5,6.0,7.7], dtype="float32")
q = model.predict(x)
print(q)

z = model.fitted()
```

```python
#Visualize the result
from mpl_toolkits.mplot3d import Axes3D  # noqa: F401 unused import
import matplotlib.pyplot as plt 

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

#plot the generated points
xs = y[:,0]
ys = y[:,1]
zs = y[:,2]
ax.scatter(xs, ys, zs)

#plot the predicted points
xx = z[:,0]
yy = z[:,1]
zz = z[:,2]
ax.plot(xx,yy,zz, c= "red")

ax.set_xlim(-1.1, 1.1)
ax.set_ylim(-1.1, 1.1)
ax.set_zlim(-1.1, 1.1)
ax.set_aspect('equal', 'box')

plt.show()
```

```python
#also plot the surface

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Make data
u = np.linspace(0, 2 * np.pi, 100)
v = np.linspace(0, np.pi, 100)
sx = np.outer(np.cos(u), np.sin(v))
sy = np.outer(np.sin(u), np.sin(v))
sz = np.outer(np.ones(np.size(u)), np.cos(v))

# Plot the surface
ax.plot_surface(sx, sy, sz, rstride=1, cstride=1, color='c', alpha=0.2, linewidth=0)

ax.scatter(xs, ys, zs, c="blue",s=20) 
ax.plot(xx,yy,zz, c= "red")

ax.set_xlim([-1,1])
ax.set_ylim([-1,1])
ax.set_zlim([-1,1])
ax.set_aspect("equal")
plt.show()
```
